```python
import pandas as pd
import warnings

warnings.filterwarnings('ignore')

mainDataFrame = pd.read_csv(r'datasets/beer_consumption.csv', sep=';')
```

```python
#correlation

mainDataFrame[['temp_avg','temp_max','temp_min','wkn','consume']].corr()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,6))
ax.set_title("Beer consume in line graph", fontsize=16)
ax.set_ylabel("Liters", fontsize=16)
ax.set_xlabel("Days", fontsize=16)
ax = mainDataFrame['consume'].plot()
```

```python
import seaborn as sns

#using boxplot

axBoxPlot = sns.boxplot(data=mainDataFrame['consume'], orient='v', width=0.2)
axBoxPlot.figure.set_size_inches(12, 6)
axBoxPlot.set_title("Beer consume in box plot")
axBoxPlot.set_ylabel("Liters", fontsize=16)
```

```python
import seaborn as sns

#comparating with boxplot

axBoxPlot = sns.boxplot(data=mainDataFrame, x='wkn',y='consume', orient='v', width=0.2)
axBoxPlot.figure.set_size_inches(12, 6)
axBoxPlot.set_title("Beer consume in box plot")
axBoxPlot.set_ylabel("Liters", fontsize=16)
```

```python
import seaborn as sns

#distribuition graph

axBoxPlot = sns.distplot(mainDataFrame['consume'])
axBoxPlot.figure.set_size_inches(12, 6)
axBoxPlot.set_title("Beer consume in box plot")
axBoxPlot.set_ylabel("Liters", fontsize=16)
```

```python
axPairPlot = sns.pairplot(mainDataFrame, y_vars='consume', x_vars=['temp_min','temp_max','temp_avg','rain','wkn'], kind='reg')
axPairPlot.fig.suptitle("Dispersion between variables", fontsize=13, y=1.05)
axPairPlot
```
